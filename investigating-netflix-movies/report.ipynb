## Import Libraries

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

## Data Preparation

- Load the data and assign to a variable as DataFrame
- Preview DataFrame

```python
netflix_df = pd.read_csv("https://raw.githubusercontent.com/cogxen/datasets/main/datacamp/netflix.csv")

# Preview
netflix_df.head()
```

- Preview the subset DataFrame where in `type` column has only "Movie"

```python
netflix_df_subset = netflix_df[netflix_df["type"] == "Movie"]

# Preview
netflix_df_subset.head()
```

- Filter only movies that was released before 1990s

```python
movies_1990s = netflix_df_subset[(netflix_df_subset["release_year"] >= 1990) & (netflix_df_subset["release_year"] < 2000)]

# Preview
movies_1990s.head()
```

- Preview the most frequent movie duration

```python
plt.hist(movies_1990s["duration"])
plt.title("Movie Durations")
plt.xlabel("Duration (mins)")
plt.ylabel("Number of Mobies")
plt.show()
```

- Filter the `movies_1990s` DataFrame and keep only "Action" movies

```python
actions = movies_1990s[movies_1990s["genre"] == "Action"]
```

- Count the number of short action movies from the 1990s

```python
# Unknown solution from the Solution tab
duration = 100

# Define a short movie as one with a duration of less than 90 minutes
short_action_movies_1990s = actions[actions["duration"] < 90]

# Count the number of short action movies
num_short_action_movies_1990s = short_action_movies_1990s.shape[0]

num_short_action_movies_1990s
```
